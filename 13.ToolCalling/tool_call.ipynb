```python
# Tool creation
import json
import requests
from datetime import date
from langchain_core.tools import tool, InjectedToolArg
from typing import Annotated

currency_details = "https://cdn.jsdelivr.net/npm/@fawazahmed0/currency-api@latest/v1/currencies.json"
conversion_url = "https://cdn.jsdelivr.net/npm/@fawazahmed0/currency-api@{today_str}/v1/currencies/{base_currency_short_form}.json"

@tool
def fetch_currency_details():
    """
    Fetches the list of all available currencies and their full names.

    This function sends an HTTP GET request to the public currency API
    endpoint and retrieves the mapping of currency short forms
    (like USD, INR, EUR) to their full currency names.

    Returns:
        str: JSON string containing currency codes and their corresponding
             currency names if the request is successful.

        str: Error message if the API request fails.
    """

    response = requests.get(currency_details)

    if response.status_code == 200:
        return json.dumps(response.json())

    return f"Request failed with status code: {response.status_code}"

@tool
def get_conversion_factor(base_currency_short_form: str, target_currency_short_form: str) -> float:
    """
    Fetch the currency conversion factor between two currencies.

    This function retrieves the latest conversion rate from the
    public currency API for converting a base currency to a target currency.

    The API endpoint dynamically uses today's date to fetch the
    most recent exchange rate.

    Args:
        base_currency_short_form (str):
            The base currency code (e.g., "usd", "inr", "eur").

        target_currency_short_form (str):
            The target currency code to which conversion is required
            (e.g., "inr", "usd", "gbp").

    Returns:
        float:
            The conversion rate between the base currency and target currency.

        float:
            Returns -1.0 if the API request fails or if the currency
            conversion rate is not found.
    """

    today_str = date.today().strftime("%Y-%m-%d")
    base_currency_short_form = base_currency_short_form.lower()
    target_currency_short_form = target_currency_short_form.lower()

    final_conversion_url = conversion_url.format(
        today_str=today_str,
        base_currency_short_form=base_currency_short_form
    )

    response = requests.get(final_conversion_url)

    if response.status_code == 200:
        return response.json().get(base_currency_short_form, {}).get(target_currency_short_form)

    return -1.00


@tool
def convert(base_currency_value: int, conversion_rate: Annotated[float, InjectedToolArg]) -> float:
    """
    Convert a base currency value into a target currency value.

    This function multiplies the base currency amount with the
    provided conversion rate to compute the final converted value.

    The conversion rate is marked with `InjectedToolArg`, meaning
    that the value will be automatically injected by the LangChain
    tool execution system rather than being provided directly by the user.

    Args:
        base_currency_value (int):
            The amount in the base currency that needs to be converted.

        conversion_rate (float):
            Exchange rate between the base and target currency.
            This value is injected automatically by the tool pipeline.

    Returns:
        float:
            Converted currency value in the target currency.
    """

    return base_currency_value * conversion_rate
```

### InjectedToolArg
Used in LangChain tools to mark arguments that should NOT be provided by the LLM directly.

Instead, the framework injects the value automatically from another tool or system.

### Annotated
From Python’s typing module.
It allows attaching metadata to a variable type.

#### Why InjectedToolArg is Powerful
Without it:  
The model might try to guess the rate:
convert(100, ???)

With InjectedToolArg:  
LangChain ensures:
conversion_rate = output of previous tool

So the agent pipeline becomes safe and deterministic.

```python
# LLM creation
llm = ChatOllama(model="qwen3-vl:4b")
```

```python
# Bind the llm with tools
llm_with_tools = llm.bind_tools([fetch_currency_details, get_conversion_factor, convert])
```

```python
messages = [
    SystemMessage(
        """
You are a helpful AI assistant that performs currency conversions using available tools.
When calling tools, always return tool calls using the tool calling format.
Do not output tool calls as JSON in the message content.

You have access to the following tools:

1. fetch_currency_details  
   - Use this tool when you need to retrieve the list of supported currencies.

2. get_converion_factor  
   - Use this tool to fetch the conversion rate between two currencies.

3. convert  
   - Use this tool to convert a currency value using the conversion rate.

Instructions:
- When a user asks for currency conversion, first call the tool that retrieves the conversion factor.
- Then use that conversion factor to call the conversion tool.
- Do not guess conversion rates.
- Always rely on the tools for accurate information.
- Provide the final answer clearly after executing the required tools.
"""
    ),

    HumanMessage(
        "What is the conversion factor between INR and USD, and based on that can you convert 10 inr to usd"
    )
]
```

```python
messages
```

```python
ai_message = llm_with_tools.invoke(messages)
messages.append(ai_message)

conversion_rate = None

for tool_call in ai_message.tool_calls:

    if tool_call["name"] == "fetch_currency_details":
        tool_message = fetch_currency_details.invoke(tool_call)

    elif tool_call["name"] == "get_conversion_factor":
        tool_message = get_conversion_factor.invoke(tool_call)
        conversion_rate = tool_message.content

    elif tool_call["name"] == "convert":
        if conversion_rate is None:
            raise ValueError("Conversion rate not available")

        tool_call["args"]["conversion_rate"] = conversion_rate
        tool_message = convert.invoke(tool_call)

    messages.append(tool_message)
```

```python
messages
```

```python
llm_with_tools.invoke(messages).content
```

```python
# Agent creation using langchain
from langchain.agents import create_agent

agent = create_agent(
    model=llm,
    tools=[fetch_currency_details, get_conversion_factor, convert],
    system_prompt="""
You are an AI assistant that performs currency conversions using tools.

Always use the provided tools to obtain currency information. 
Never guess or assume exchange rates.

Available tools:
1. fetch_currency_details – retrieves the list of supported currencies.
2. get_conversion_factor – returns the conversion rate between two currencies.
3. convert – converts an amount using a provided conversion rate.

Workflow for currency conversion:
1. Identify the source currency, target currency, and amount.
2. Call get_conversion_factor to retrieve the conversion rate.
3. Use the returned rate to call convert with the amount.
4. Return the final converted value to the user.

Rules:
- Always use tools when performing conversions.
- Do not output tool calls as JSON in the message content.
- Use the tool-calling interface provided by the system.
- Only provide the final answer after the required tools have been executed.
"""
)
```

```python
response = agent.invoke({
    "messages": [
        {"role": "user", "content": "Convert 100 USD to INR"}
    ]
})

print(response)
```
