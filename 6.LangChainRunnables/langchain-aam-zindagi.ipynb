```python
import random

class NakliLLM:
    
    def __init__(self):
        print("LLM created")

    def predict(self, prompt):
        response_list = [
            'Delhi is the capital of India',
            'IPL is a cricket league',
            'AI stands for Artificial Intelligence'
        ]
        return {'response': random.choice(response_list)}
```

```python
class NakliPromptTemplate:
    def __init__(self, template, input_variables):
        self.template = template
        self.input_variables = input_variables

    def format(self, input_dict):
        return self.template.format(**input_dict)  
```

```python
template = NakliPromptTemplate(
    template="Write a {length} poem about the {topic}",
    input_variables=['length', 'topic']
)
```

```python
prompt = template.format({
    'length': 'short',
    'topic': 'india'
})
```

```python
# LLM creation
llm = NakliLLM()
```

```python
llm.predict(prompt=prompt)
```

## Creating chains to automate the reusable task
e.g calling the llm after getting fromatted prompt

```python
class NakliLLMChain:
    def __init__(self, llm, template):
        self.llm: NakliLLM  = llm
        self.template: NakliPromptTemplate = template

    def run(self, input_dict):
        final_prompt = self.template.format(input_dict)
        result = self.llm.predict(final_prompt)

        return result['response']
```

```python
# Tempalte creation
template = NakliPromptTemplate(
    template='Write a {length} poem about {topic}',
    input_variables=['length', 'topic']
)
```

```python
# LLM creation
llm = NakliLLM()
```

```python
# Chain creation
chain = NakliLLMChain(llm=llm, template=template)
```

```python
# Invocation
chain.run({
    'length': 'short',
    'topic': 'india'
})
```
