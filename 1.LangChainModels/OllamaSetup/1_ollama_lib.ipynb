```python
# Giving multimodal input to model (passing immage to the model)
# 
import base64
import os
import ollama

# Using os.path.join for OS-independent path handling
# For Jupyter notebooks, use os.getcwd() to get the current working directory
image_path = os.path.join(os.getcwd(), "..", "images", "wp3753002.jpg")
image_path = os.path.abspath(image_path)

with open(image_path, "rb") as image_file:
    image_bytes = image_file.read()

image_64 = base64.b64encode(image_bytes).decode("utf-8") # Encoding the image using Base64

response = ollama.generate(model="gemma3:4b", images=[image_64], prompt="Give caption for image")
print(response.response)
```

```python
# Giving multimodal input to model (passing immage to the model)
# llama3.2:latest --> Text bases model. Here the mdoel won't be able to interpret the image
import base64
import os
import ollama

# Using os.path.join for OS-independent path handling
# For Jupyter notebooks, use os.getcwd() to get the current working directory
image_path = os.path.join(os.getcwd(), "..", "images", "wp3753002.jpg")
image_path = os.path.abspath(image_path)

with open(image_path, "rb") as image_file:
    image_bytes = image_file.read()

image_64 = base64.b64encode(image_bytes).decode("utf-8") # Encoding the image using Base64

response = ollama.generate(model="llama3.2:latest", images=[image_64], prompt="Give caption for image")
print(response.response)
```

```python
# System instructions
response = ollama.generate(
    model="gemma3:4b",
    prompt="Why does the moon glow?",
    system="Your are a funny assistant, you explain things in a funny way"
)
print(response.response)
```

```python
# Changing the option parameters
response = ollama.generate(
    model="gemma3:4b",
    prompt="Why is ocean blue?",
    system="You are a concise assistant, provide brief and to-the-point answers.",
    options={"top_p": 0.5, "temperature": 0.3, "top_k":45}
)
print(response.response)
```
